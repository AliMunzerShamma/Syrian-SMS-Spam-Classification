```python
import pandas as pd
from pandas import DataFrame
import csv
import re
```

```python
data = []
pattern = r"(.)*;"
pattern2 = r"[!”\'\"#$%&’()*+,-./:;<=>؟?@[\]^_`{|}~]"
with open('spam14.csv','r',encoding="utf16",newline='') as fd:
    Reader = csv.reader(fd)
    for row in Reader:
        newRow = re.sub(pattern, "", str(row))
        newRow = re.sub(pattern2, "", newRow)
        if newRow != "":
            data.append(newRow)
        
data.remove('Received messages')
```

```python
Class_data = []
for i in range(len(data)):
    Class_data.append("spam")
    
#Class_data.append("spam")
#Class_data.append("spam")
```

```python
with open('Data_spam.csv','a',encoding="utf-8-sig",newline='') as fd:
    writer = csv.writer(fd)
    for i in range(len(data)):
        writer.writerow([data[i], Class_data[i]])
```
